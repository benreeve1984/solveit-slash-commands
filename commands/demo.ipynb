# SolveIt Slash Commands — Demo

A quick tour of the six built-in slash commands.
Make sure the CRAFT setup block has run first (restart your kernel if needed).

## /compact

Hides low-value messages from AI context to free up your context window.
No args needed — it analyses the whole dialog.

```
/compact
```

## /repro

Adds a new code cell with a minimal standalone bug reproduction.
Describe the bug after the command.

```
/repro the list comprehension raises IndexError on empty input: result = [items[0] for items in data]
```

## /tidy

Proposes how to reorder and clean up the dialog for sharing.
Doesn't touch anything — just gives you a plan.

```
/tidy
```

## /simplify

Proposes code simplifications following fastai style.
Never edits cells — you apply what you like.

```
/simplify the data loading section
```

## /craft

Generates a CRAFT.ipynb capturing this project's context and setup.
Non-destructive — outputs a code block you can save.

```
/craft
```

## /new_command

Meta-command: creates a new slash command interactively.
Give it a name and description.

```
/new_command validate run all assertions and report failures before committing
```
