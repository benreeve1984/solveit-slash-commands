OK, I've read through the docs thoroughly. Here's my analysis of how slash commands could work in SolveIt:

**The core design question is: who expands the template — Python or the AI?**

### Approach A: AI Tool (cleanest UX)

User types naturally in a prompt:
```
/new_feature add a green button labelled 'go'
```

A CRAFT note tells the AI: "When a prompt starts with `/`, call the `slash_cmd` tool to load the template." The tool reads `~/commands/new_feature.md`, substitutes variables, and returns the expanded instructions. The AI then follows them.

- ✅ Clean, natural syntax — just type `/cmd args`
- ✅ AI can interpret args intelligently (no rigid parsing needed)
- ❌ Relies on AI recognizing the convention (one extra tool-call round trip)

### Approach B: Variable Injection (deterministic)

User writes in a prompt:
```
$`sc("new_feature add a green button labelled 'go'")`
```

A CRAFT-defined `sc()` function reads the template, substitutes args, and injects the expanded text into the AI's context automatically — before the AI even sees it.

- ✅ Deterministic — no AI interpretation needed
- ✅ Works identically across models
- ❌ Clunkier syntax (Python string inside `$` backticks)

### Approach C: Hybrid

Combine both — `$`sc(...)`` for the expansion, plus a CRAFT note so the AI understands what happened. Or even: a CRAFT note that says "if the user just writes `/cmd args`, call the tool; if they use `$`sc(...)`` the expansion is already in context."

```markdown

description: Add a new feature to the project

2. Plan the implementation  
3. Write the code
```

Stored in `~/commands/new_feature.md` (paralleling the existing `~/skills/` convention).

IPython has a hook called an **input transformer** — it's a function that receives the raw lines of a code cell *before* IPython tries to parse them as Python. You can use it to rewrite input into valid Python.

So the flow would be:

1. User types `/new_feature add a green button labelled 'go'` in a **code cell**
2. The input transformer sees it starts with `/`, and rewrites it to something like `sc("new_feature", "add a green button labelled 'go'")`
3. IPython executes that rewritten Python normally
4. `sc()` reads the template, expands variables, and calls `add_prompt()`

The key bit is registering the transformer. It's just:

```python
from IPython import get_ipython

def slash_transform(lines):
    # check if first line starts with '/'
    # if so, rewrite to sc() call
    # otherwise return lines unchanged
    return lines

get_ipython().input_transformers_cleanup.append(slash_transform)
```

That registration would go in your CRAFT code block so it's always active. The `sc()` function itself handles the template loading and variable expansion.

Yes! dialoghelper can toggle AI visibility programmatically using update_msg(). Use the skipped parameter:

from dialoghelper import *

# Hide a message from AI context
await update_msg(msg_id, skipped=1)

# Unhide a message
await update_msg(msg_id, skipped=0)


You can combine this with find_msgs() to bulk toggle visibility based on patterns:

msgs = await find_msgs("some pattern")
for m in msgs:
    await update_msg(m['id'], skipped=1)  # hide all matching
